```python
X = Stock.values
train_size = int(len(X) * 0.90)
train, test = X[0:train_size], X[train_size:len(X)]
print('Observations: %d' % (len(X)))
print('Training Observations: %d' % (len(train)))
print('Testing Observations: %d' % (len(test)))
```

```python
Stock_Train=Stock[0:2100]
Stock_Validate=Stock[2101:2300]
Stock_Test=Stock[2300:2477]
print(Stock_Train,Stock_Validate,Stock_Test)
```

```python
print(f"the size of the traing set size is ",Stock_Train.shape)
print(f"the size of the validate set size is ",Stock_Validate.shape)
print(f"the size of the testing set size is ",Stock_Test.shape)
```

```python
from statsmodels.tsa.arima.model import ARIMA
import sklearn.metrics as metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(Stock_Train, order=(1,0,0))
result = model.fit()
#print result
start=1
end=len(Stock_Train)
pred=result.predict(start=start,end=end,typ='levels')
#print pred
x=np.sqrt(metrics.mean_squared_error(Stock_Train,pred))

start=len(Stock_Train)
end=len(Stock_Train)+len(Stock_Validate)-1
pred=result.predict(start=start,end=end,typ='levels')
#print pred
y=np.sqrt(metrics.mean_squared_error(Stock_Validate,pred))
print("The ARIMA model (1,0,0)")
print(f"Squred error for the train",x)
print(f"Squred error for the validate",y)
```

```python
#we can split the time series into two contiguous sequences. 
#We can then calculate the mean and variance of each group of numbers and compare the values.
X = Stock.values
split = round(len(X) / 2)
X1, X2 = X[0:split], X[split:]
mean1, mean2 = X1.mean(), X2.mean()
var1, var2 = X1.var(), X2.var()
print('mean1=%f, mean2=%f' % (mean1, mean2))
print('variance1=%f, variance2=%f' % (var1, var2))
```

```python
#Running this example shows that the mean and variance values are different, but in the same ball-park^^^
```
